```python
# load packages
import numpy as np
import torch
from sklearn.metrics import brier_score_loss, log_loss, accuracy_score, precision_score, recall_score, f1_score
from torchcp.classification.scores import  SAPS
from torchcp.classification.predictors import ClassWisePredictor
import pandas as pd

from model.DeepLOB import deeplob
from utils.torch_dfs import LobDataset
from utils.constants import DEVICE
from helpers.result_datahandler import ResultsDataHandler
```

```python
batch_size = 64


dec_data = np.loadtxt('data/input/Train_Dst_NoAuction_DecPre_CF_7.txt')

dec_cal = dec_data[:, int(np.floor(dec_data.shape[1] * 0.8)):int(np.floor(dec_data.shape[1] * 0.975))]
dataset_cal = LobDataset(data=dec_cal, k=4, num_classes=3, T=100)
cal_loader = torch.utils.data.DataLoader(dataset=dataset_cal, batch_size=batch_size, shuffle=False)

print('Calibration Data Shape:', dataset_cal.x.shape, dataset_cal.y.shape)

del dec_cal, dec_data, dataset_cal

dec_test1 = np.loadtxt('data/input/Test_Dst_NoAuction_DecPre_CF_7.txt')
dec_test2 = np.loadtxt('data/input/Test_Dst_NoAuction_DecPre_CF_8.txt')
dec_test3 = np.loadtxt('data/input/Test_Dst_NoAuction_DecPre_CF_9.txt')
dec_test = np.hstack((dec_test1, dec_test2, dec_test3))

dataset_test = LobDataset(data=dec_test, k=4, num_classes=3, T=100)
test_loader = torch.utils.data.DataLoader(dataset=dataset_test, batch_size=batch_size, shuffle=False)


print('Test Data Shape:',dataset_test.x.shape, dataset_test.y.shape)

del dec_test, dec_test1, dec_test2, dec_test3, dataset_test
```

```python
model = deeplob(y_len = 3)
model_path = 'model/best_val_model_pytorch'

model = torch.load(model_path,  map_location=torch.device(DEVICE))
```

```python
alpha = 0.1
fun = 'SAPS'
params = res_maxuni.hyperparam_df

best_model_params = params[(params['Alpha'] == str(alpha))&(params['Function'] == fun)]

best_predictor = ClassWisePredictor(score_function=SAPS(best_model_params['Lambda'].iloc[0]), model=model, temperature=best_model_params['Best_Temperature'].iloc[0])
best_predictor.calibrate(cal_loader, alpha)
```

```python
best_predictor.evaluate(test_loader)
```

```python
# Placeholder lists for true labels and probabilities
true_labels = []
base_model_probabilities = []
conformal_model_probabilities = []

# Placeholder lists for predictions
base_model_pred_flattened = []
conformal_pred_labels = []

# Additional placeholders for filtering by size 1 prediction sets
filtered_true_labels = []
filtered_conformal_pred_labels = []
total_conformal_sets = 0
size_1_conformal_sets = 0

# Populate the true_labels and predictions
for inputs, targets in test_loader:
    # Move to GPU
    inputs, targets = inputs.to(DEVICE, dtype=torch.float), targets.to(DEVICE, dtype=torch.int64)
    
    # Collect true labels
    true_labels.extend(targets.tolist())
    
    # Base Model Predictions
    base_probs = model(inputs).detach()
    base_model_probabilities.extend(base_probs.cpu().numpy())
    _, predictions = torch.max(base_probs, 1)
    base_model_pred_flattened.extend(predictions.tolist())
    
    # Conformal Model Predictions
    pred_sets = best_predictor.predict(inputs)
    conformal_probs = best_predictor.predict_probabilities(inputs)
    conformal_model_probabilities.extend(conformal_probs.cpu().numpy())
    
    # Select the prediction with the highest probability or fallback to 1
    for i, (pred_set, probs) in enumerate(zip(pred_sets, conformal_probs)):
        total_conformal_sets += 1
        
        if len(pred_set) == 1:
            # Size-1 prediction set
            size_1_conformal_sets += 1
            conformal_pred_labels.append(pred_set[0])
            filtered_true_labels.append(targets[i].item())
            filtered_conformal_pred_labels.append(pred_set[0])
        else:
            # Move probs to CPU and convert to NumPy before applying argmax
            if len(pred_set) > 0:
                highest_prob_label = pred_set[np.argmax(probs[pred_set].cpu().numpy())]
                conformal_pred_labels.append(highest_prob_label)
            else:
                conformal_pred_labels.append(1)  # Fallback to 1 if no valid prediction

# Convert to numpy arrays
true_labels = np.array(true_labels)
base_model_probabilities = np.array(base_model_probabilities)
conformal_model_probabilities = np.array(conformal_model_probabilities)
base_model_pred_flattened = np.array(base_model_pred_flattened)
conformal_pred_labels = np.array(conformal_pred_labels)
filtered_true_labels = np.array(filtered_true_labels)
filtered_conformal_pred_labels = np.array(filtered_conformal_pred_labels)

# Calculate Brier score for each class and average them for the base model
brier_base = np.mean([
    brier_score_loss(true_labels == i, base_model_probabilities[:, i])
    for i in range(base_model_probabilities.shape[1])
])

# Calculate Brier score for each class and average them for the conformal model
brier_conformal = np.mean([
    brier_score_loss(true_labels == i, conformal_model_probabilities[:, i])
    for i in range(conformal_model_probabilities.shape[1])
])

# Calculate log loss for base model
log_loss_base = log_loss(true_labels, base_model_probabilities)

# Calculate log loss for conformal model
log_loss_conformal = log_loss(true_labels, conformal_model_probabilities)

# Calculate model-level accuracy
accuracy_base_model = accuracy_score(true_labels, base_model_pred_flattened)
accuracy_conformal_model = accuracy_score(true_labels, conformal_pred_labels)

# Calculate model-level accuracy for filtered conformal model
accuracy_filtered_conformal_model = accuracy_score(filtered_true_labels, filtered_conformal_pred_labels) if len(filtered_true_labels) > 0 else np.nan

# Percentage of size-1 sets
percentage_size_1 = (size_1_conformal_sets / total_conformal_sets) * 100

# Function to calculate metrics for a given label
def calculate_metrics(true, pred, label):
    accuracy = accuracy_score(true == label, pred == label)
    precision = precision_score(true, pred, labels=[label], average='macro', zero_division=0)
    recall = recall_score(true, pred, labels=[label], average='macro', zero_division=0)
    f1 = f1_score(true, pred, labels=[label], average='macro', zero_division=0)
    return accuracy, precision, recall, f1

# Collect metrics in dictionaries
metrics_base = {'Label': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}
metrics_conformal = {'Label': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}
metrics_filtered_conformal = {'Label': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}

for label in [0, 1, 2]:
    # Metrics for Base Model
    accuracy, precision, recall, f1 = calculate_metrics(true_labels, base_model_pred_flattened, label)
    metrics_base['Label'].append(label)
    metrics_base['Accuracy'].append(accuracy)
    metrics_base['Precision'].append(precision)
    metrics_base['Recall'].append(recall)
    metrics_base['F1-Score'].append(f1)
    
    # Metrics for Unfiltered Conformal Model
    accuracy, precision, recall, f1 = calculate_metrics(true_labels, conformal_pred_labels, label)
    metrics_conformal['Label'].append(label)
    metrics_conformal['Accuracy'].append(accuracy)
    metrics_conformal['Precision'].append(precision)
    metrics_conformal['Recall'].append(recall)
    metrics_conformal['F1-Score'].append(f1)
    
    # Metrics for Filtered Conformal Model
    if len(filtered_true_labels) > 0:  # Ensure there are filtered predictions
        accuracy, precision, recall, f1 = calculate_metrics(filtered_true_labels, filtered_conformal_pred_labels, label)
    else:
        accuracy = precision = recall = f1 = np.nan

    metrics_filtered_conformal['Label'].append(label)
    metrics_filtered_conformal['Accuracy'].append(accuracy)
    metrics_filtered_conformal['Precision'].append(precision)
    metrics_filtered_conformal['Recall'].append(recall)
    metrics_filtered_conformal['F1-Score'].append(f1)

# Convert dictionaries to DataFrames for display
df_metrics_base = pd.DataFrame(metrics_base)
df_metrics_conformal = pd.DataFrame(metrics_conformal)
df_metrics_filtered_conformal = pd.DataFrame(metrics_filtered_conformal)

# Print the percentage of size-1 sets
print(f"Percentage of prediction sets of size 1: {percentage_size_1:.2f}, Alpha: {alpha}%")
print('\n')
# Print Brier score and Log loss
print(f"Brier Score (Base Model): {brier_base:.4f}")
print(f"Brier Score (Conformal Model): {brier_conformal:.4f}")
print('\n')
print(f"Log Loss (Base Model): {log_loss_base:.4f}")
print(f"Log Loss (Conformal Model): {log_loss_conformal:.4f}")
print('\n')
# Print model-level accuracy
print(f"Model-Level Accuracy (Base Model): {accuracy_base_model:.4f}")
print(f"Model-Level Accuracy (Conformal Model): {accuracy_conformal_model:.4f}")
print(f"Model-Level Accuracy (Filtered Conformal Model): {accuracy_filtered_conformal_model:.4f}")

# Display the metrics for the Base Model
print("\nBase Model Metrics:")
print(df_metrics_base.to_string(index=False))

# Display the metrics for the Unfiltered Conformal Model
print("\nConformal Model Metrics (Unfiltered):")
print(df_metrics_conformal.to_string(index=False))

# Display the metrics for the Filtered Conformal Model
print("\nFiltered Conformal Model Metrics (Only considering size-1 sets):")
print(df_metrics_filtered_conformal.to_string(index=False))
```
